## HW1: Linear Regression using Gradient Descent
In hw1, you need to implement linear regression by using only numpy, then train your implemented model by the provided dataset and test the performance with testing data

Please note that only **NUMPY** can be used to implement your model, you will get no points by simply calling sklearn.linear_model.LinearRegression

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

## Load data

```python
train_df = pd.read_csv("train_data.csv")
x_train, y_train = train_df['x_train'], train_df['y_train']
```

```python
plt.plot(x_train, y_train, '.')
```

## Train your model 
### Pseudo code:
1. Random initialize the weights, intercepts of the linear model

**for i in range(iteration)**

    2. Feed foward the training data into the model, get the output prediction
    3. Calculating training loss by Mean Square Error of predcition and ground truth data
    4. Calculating the gradients
    5. Updating the weights and intercepts by the gradients * learning rate 
    
**End of training**

```python
iteration = 300
lr = 1e-2
w = np.random.randn()
b = np.random.randn()
L = []

# Calculate weight and intercept by using gradient descent
for i in range(iteration):
    loss = ((w * x_train + b - y_train)**2).mean()
    L.append(loss)
    delta_w = (2*w*x_train*x_train + 2*b*x_train - 2*x_train*y_train).mean()
    delta_b = (2*w*x_train - 2*y_train).mean() + 2*b
    w -= lr*delta_w
    b -= lr*delta_b

# Draw the estimated curve and ground truth of training data
plt.plot(x_train, w*x_train+b)
plt.plot(x_train, y_train, ".")
plt.show()

# Draw learning curve
plt.plot(L)
plt.show()
```

## Test the performance on the testing data
Inference the test data (x_test) by your model and calculate the MSE of (y_test, y_pred)

```python
test_data = pd.read_csv("test_data.csv")
x_test, y_test = test_data['x_test'], test_data['y_test']
```

```python
# Return the predict value
def model(x):
    return w*x + b

# Draw predict value and  ground truth of testing data
y_pred = model(x_test)
plt.plot(x_test, y_pred)
plt.plot(x_test, y_test, '.')
plt.show()

# Calculate mean square error
error = ((y_test-y_pred)**2).mean()

print('Mean square error is ', error)
print('Weight is ', w)
print('Intercept is ', b)
```

## What’s the difference between Gradient Descent, Mini-Batch Gradient Descent, and Stochastic Gradient Descent?
Gradient Descent: Use all elements in training set to calculate loss function in one iteration, then take the derivative
 of loss funtion, after that the weight and intercept can be updated by using learning rate and gradient.

Mini-Batch Gradient Descent: Select k  elements (k is smaller than the size of training set) from the training set 
randomly to calculate loss function in one iteration, then take the derivative
 of loss funtion, after that the weight and intercept can be updated by using learning rate and gradient.

Stochastic Gradient Descent: Select 1 elements from the training set randomly to calculate loss function in one 
iteration, then take the derivative
 of loss funtion, after that the weight and intercept can be updated by using learning rate and gradient.
